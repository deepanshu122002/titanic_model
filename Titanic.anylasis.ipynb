```python
df.replace({'Sex':{'male':1,'female':0}, 'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
```

```python
A=df.drop(columns=['Name','Ticket','PassengerId','Survived'],axis=1)
```

```python
B = df['Survived']
```

```python
A
```

```python
B
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
A_train,A_test,B_train,B_test = tts(A, B, test_size=0.2,random_state=2)
```

```python
print(A_train.shape,A.shape, A_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression as lg
```

```python
model = lg()
```

```python
model.fit(A_train,B_train)
```

```python
A_train_prediction = model.predict(A_train)
```

```python
print(A_train_prediction)
```

```python
from sklearn.metrics import accuracy_score as acs
```

```python
training_data_accuracy = acs(B_train, A_train_prediction)
```

```python
print('Accuracy_score_training_data : ',training_data_accuracy)
```

```python
A_test_prediction = model.predict(A_test)
```

```python
print(A_test_prediction)
```

```python
test_data_accuracy = acs(B_test, A_test_prediction)
print('Accuracy_score_test_data : ',test_data_accuracy)
```
